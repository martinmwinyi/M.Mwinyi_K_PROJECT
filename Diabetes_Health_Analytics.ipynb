STEP 6: DATA VISUALIZATION

```python
plt.hist(df['Age'], bins=20)
plt.title("Age Distribution of Patients")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
df.groupby('Age')['DiabetesPedigreeFunction'].mean().plot(kind='line')
plt.title("Disease Risk by Age")
plt.xlabel("Age")
plt.ylabel("Average Disease Risk")
plt.grid()
plt.show()
```

```python
df.head()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Glucose'], df['Outcome'], alpha=0.6)
plt.title("Glucose vs  Disease Outcome")
plt.xlabel("Glucose Level")
plt.ylabel("Outcome")
plt.grid()
plt.show()
```

```python
df.groupby('Age')['Outcome'].mean().plot()
plt.title("Health Risk Trend by Age")
plt.xlabel("Age")
plt.ylabel("Risk Level")
plt.grid()
plt.show()
```

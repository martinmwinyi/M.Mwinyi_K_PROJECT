STEP 6: DATA VISUALIZATION

1) Age Distribution of Patients

```python
plt.hist(df['Age'], color='skyblue')
plt.title("Age Distribution of Patients")
plt.xlabel("Age")
plt.ylabel("Count")
plt.grid(linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

2) Average Diabetes Risk

```python
df.groupby('Age')['DiabetesPedigreeFunction'].mean().plot(
    kind='line',
    linestyle=':',
    marker='o',
    markerfacecolor='black',
    linewidth=2,
    color='red',
    label= 'Diabetes Risk Changes'
)

plt.title("Average Diabetes Risk")
plt.xlabel("Age")
plt.ylabel("Average Diabetes Pedigree Function")
plt.legend(loc='upper left')
plt.grid(axis='y', linestyle=':', alpha=0.6, color='green')
plt.tight_layout()
plt.show()
```

3) Glucose Level Vs Diabetes Outcome

```python
plt.figure(figsize=(8,5))
plt.scatter(
    df['Glucose'], 
    df['Outcome'], 
    c=df['Glucose'], 
    cmap='hot', 
    alpha=0.5
)

plt.title("Glucose Levels vs Diabetes Diagnosis")
plt.xlabel("Glucose Level")
plt.ylabel("Outcome")
plt.colorbar(label='Glucose Level')
plt.grid(linestyle=':')
plt.show()
```

4) Health Risk Trend by Age 

```python
df.groupby('Age')['Outcome'].mean().plot()
plt.title("Health Risk Trend by Age")
plt.xlabel("Age")
plt.ylabel("Risk Level")
plt.grid(True, linestyle='dashdot', color='green')
plt.show()
```
